# 缠论数据图表

```python
from cl_v2 import cl
from cl_v2 import kcharts
from cl_v2 import exchange_futu
from prettyprinter import cpprint
import time, datetime
import pandas as pd

```

```python
code = 'SH.000001'
frequencys = ['d']
start_date = None
end_date = None
exchange = exchange_futu.ExchangeFutu()
```

```python
renders = []
klines = {}
for f in frequencys:
    _ks = exchange.klines(code, frequency=f, start_date = start_date, end_date = end_date)
#     _ks = kline_ks0:988:]
    print(f, '获取K线数据量：', len(_ks))
    klines[f] = _ks

cl_datas = cl.batch_cls(code, klines)
for cd in cl_datas:
    title = '%s - 【%s】 周期数据图表' % (code, cd.frequency)
    renders.append(kcharts.render_charts(title, cd))
```
